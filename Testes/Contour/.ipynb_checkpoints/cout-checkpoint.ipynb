```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
xlist = np.linspace(0.0, 3.0, 100)
ylist = np.linspace(0.0, 2*np.pi, 100)
X, Y = np.meshgrid(xlist, ylist)
```

```python
t,phi_1_r,phi_2_r,phi_3_r,phi_1_i,phi_2_i,phi_3_i= np.loadtxt("phi2.dat",delimiter="\t",unpack=True)
```

```python
t,phi_1,phi_2,phi_3= np.loadtxt("phi_abs.dat",delimiter="\t",unpack=True)
```

```python


fig,ax=plt.subplots()
def animate(frame_num):
    time = frame_num
    Z = phi_1_r[time]*np.cos(17.5*X+5*Y)-phi_1_i[time]*np.sin(17.5*X+5*Y)+phi_2_r[time]*np.cos(-7.5*X-2.5*Y)-phi_2_i[time]*np.sin(-7.5*X-2.5*Y)+phi_3_r[time]*np.cos(-10*X-2.5*Y)-phi_3_i[time]*np.sin(-10*X-2.5*Y)-5.0*X
    ax.clear()
    ax.contourf(X, Y, Z)
    plt.title(r't = %i' % time )
    #fig.colorbar(cp)
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)



```

```python
anim = FuncAnimation(fig, animate, frames=100)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()     
```
