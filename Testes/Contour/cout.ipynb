```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
xlist = np.linspace(0.0, 3.0, 100) #cria uma lista de valores entre 0 e 3 com 100 pontos
ylist = np.linspace(0.0, 2*np.pi, 100)#cria uma lista de pontos entre 0 e 2pi com 100 pontos
X, Y = np.meshgrid(xlist, ylist) # criar um grid, foi pro meu problema
```

```python
t,phi_1_r,phi_2_r,phi_3_r,phi_1_i,phi_2_i,phi_3_i= np.loadtxt("phi2.dat",delimiter="\t",unpack=True) #importa dados
```

```python
t,phi_1,phi_2,phi_3= np.loadtxt("phi_abs.dat",delimiter="\t",unpack=True)
```

```python


fig,ax=plt.subplots(1,2) #cria as duas figuras
fig.set_size_inches(20*0.393, 10*0.393) # esse fatir 0.393 Ã© p converter polegadas p cm """

time = 0 #tempo 0
#calcula o z que vai ser plotado no x,y,z
Z = phi_1_r[time]*np.cos(17.5*X+5*Y)-phi_1_i[time]*np.sin(17.5*X+5*Y)+phi_2_r[time]*np.cos(-7.5*X-2.5*Y)-phi_2_i[time]*np.sin(-7.5*X-2.5*Y)+phi_3_r[time]*np.cos(-10*X-2.5*Y)-phi_3_i[time]*np.sin(-10*X-2.5*Y)-5.0*X**2
#faz o plot
h = ax[0].contourf(X,Y,Z,cmap="jet")
fig.colorbar(h)
pote = (phi_1_r+phi_1_i+phi_2_r+phi_2_i+phi_3_r+phi_3_i)
#cria uma lista entre 0 e 500 com 501 pontos
tempo = np.linspace(0,500,501)

#plota o primeiro ponto da linha
line, =  ax[1].plot(tempo,pote)
#funcao pra fazer a animacao
def animate(frame_num):
    time = frame_num #define o frame/indice da lista dos dados que vai ser puxado
    #caclula z
    Z = phi_1_r[time]*np.cos(17.5*X+5*Y)-phi_1_i[time]*np.sin(17.5*X+5*Y)+phi_2_r[time]*np.cos(-7.5*X-2.5*Y)-phi_2_i[time]*np.sin(-7.5*X-2.5*Y)+phi_3_r[time]*np.cos(-10*X-2.5*Y)-phi_3_i[time]*np.sin(-10*X-2.5*Y)-np.cos(X)
    #Z = phi_1[time]*np.sin(17.5*X)*np.cos(5*Y-1.31e-3*time)+phi_2[time]*np.sin(-7.5*X)*np.cos(-2.5*Y-1.31e-3*time)+phi_3[time]*np.sin(-10.0*X)*np.cos(-2.5*Y-1.31e-3*time)
    #limpa a tela pro proximo plot
    ax[0].clear()
    ax[0].set_title(r't = %i' % time )
    #faz o plot
    ax[0].contourf(Y, X, Z,cmap="jet")
    #plota a linha
    line.set_data(tempo[:time],pote[:time])
    line.axes.axis([0,500,-1,4])
    




```

```python
anim = FuncAnimation(fig, animate, frames=500,interval=100) #faz a animacao, pode mexer nos numeros pra ficar melhor
video = anim.to_html5_video() # faz o video pra ver no site
html = display.HTML(video) #mostra o video aqui
display.display(html) 
plt.close()     #fecha o plot pra salvar
anim.save("fig.gif") # salva o gif
```
